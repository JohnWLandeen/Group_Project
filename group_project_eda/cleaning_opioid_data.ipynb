```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Load file
opioid_df = pd.read_csv("opioid.csv")
opioid_df
```

```python
# Drop 'Unnamed:0'
opioid_df = opioid_df.drop(columns=["Unnamed: 0"])
opioid_df
```

```python
# Data check
opioid_df.dtypes
```

```python
# Convert year to string
opioid_df["Year"] = opioid_df["Year"].astype(str)
opioid_df
```

```python
# Data type check
opioid_df.dtypes
```

```python
# 2014 sum
df_14 = opioid_df[opioid_df['Year'].str.contains("2014")]
df_14.head()
```

```python
# Total US numbers in 2014
us_14 = df_14.sum(axis = 0, skipna = True)
us_14
```

```python
# Check data type
us_14.dtype
```

```python
# 2015 sum
df_15 = opioid_df[opioid_df['Year'].str.contains("2015")]
df_15.head()
```

```python
# Total US numbers in 2015
us_15 = df_15.sum(axis = 0, skipna = True)
us_15
```

```python
# Keep raw number
us_15 = us_15["Number of Deaths"]
us_15
```

```python
# check data type
us_15.dtype
```

```python
# 2016 sum
df_16 = opioid_df[opioid_df['Year'].str.contains("2016")]
df_16.head()
```

```python
# Total US numbers in 2016
us_16 = df_16.sum(axis = 0, skipna = True)
us_16
```

```python
# Keep raw number
us_16 = us_16["Number of Deaths"]
us_16
```

```python
# 2017 sum
df_17 = opioid_df[opioid_df['Year'].str.contains("2017")]
df_17.head()
```

```python
# Total US numbers in 2017
us_17 = df_17.sum(axis = 0, skipna = True)

# Keep raw number
us_17 = us_17["Number of Deaths"]
us_17
```

```python
# 2018 sum
df_18 = opioid_df[opioid_df['Year'].str.contains("2018")]
df_18.head()
```

```python
# Total US numbers in 2018
us_18 = df_18.sum(axis = 0, skipna = True)

# Keep raw number
us_18 = us_18["Number of Deaths"]
us_18
```

```python
# 2019 sum
df_19 = opioid_df[opioid_df['Year'].str.contains("2019")]
df_19.head()
```

```python
# Total US numbers in 2018
us_19 = df_19.sum(axis = 0, skipna = True)

# Keep raw number
us_19 = us_19["Number of Deaths"]
us_19
```

```python
# 2020 sum
df_20 = opioid_df[opioid_df['Year'].str.contains("2020")]
df_20
```

```python
# Total US numbers in 2018
us_20 = df_20.sum(axis = 0, skipna = True)

# Keep raw number
us_20 = us_20["Number of Deaths"]
us_20
```

```python
# create dataframe
years = [["2014", us_14], ["2015", us_15], ['2016', us_16], ['2017', us_17], ['2018', us_18], ['2019', us_19], ['2020', us_20]]

us_data = pd.DataFrame(years, columns=["Year", "Number of Deaths"])

us_data
```

```python
# insert 
us_data.insert(1, "State", "Total U.S.", True)
us_data
```

```python
# Check data types
us_data.dtypes
```

```python
opioid_df
```

```python
# Merge into opioid_df
all_us_opioid_deaths = pd.merge(us_data, opioid_df, on=['Year', 'State', 'Number of Deaths'], how='outer')
all_us_opioid_deaths
```

```python
# check data types
all_us_opioid_deaths.dtypes
```

```python
# save as csv
all_us_opioid_deaths.to_csv('all_us_opioid_deaths.csv')
```
