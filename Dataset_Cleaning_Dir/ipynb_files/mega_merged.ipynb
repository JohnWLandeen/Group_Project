```python
import pandas as pd
import numpy as np
```

```python
# Load legality_merged_df
legality_merged_df = pd.read_csv("../datasets/legality_merged.csv")
legality_merged_df
```

```python
# Drop first column
legality_merged_df = legality_merged_df.drop(columns=["Unnamed: 0"])
legality_merged_df
```

```python
legality_merged_df.dtypes
```

```python
# Load percap_data
percap_data = pd.read_csv("../finalized_datasets/percap_data.csv")
percap_data
```

```python
percap_data = percap_data.drop(columns=['Unnamed: 0'])
percap_data
```

```python
mega_merged = pd.merge(percap_data, legality_merged_df, on=['State'], how='outer')
mega_merged
```

```python
# Replace NaN with 0
mega_merged = mega_merged.fillna(0)
mega_merged
```

```python
# Convert to int
mega_merged['Years legal (rec)'] = mega_merged['Years legal (rec)'].astype(int)
mega_merged['Years legal (med)'] = mega_merged['Years legal (med)'].astype(int)
mega_merged['Recreational'] = mega_merged['Recreational'].astype(int)
mega_merged['Medical'] = mega_merged['Medical'].astype(int)
mega_merged['Rec Year Legalized'] = mega_merged['Rec Year Legalized'].astype(int)
mega_merged['Med Year Legalized'] = mega_merged['Med Year Legalized'].astype(int)
mega_merged
```

```python
mega_merged.to_csv('../finalized_datasets/mega_merged.csv')
```
