```python
import pandas as pd
import numpy as np
```

```python
# Load revised_percap_data
revised_percap_data = pd.read_csv("../finalized_datasets/revised_percap_data.csv")
revised_percap_data
```

```python
revised_percap_data.dtypes
```

```python
revised_percap_data["Population"] = revised_percap_data["Population"].str.replace(',','')
revised_percap_data["Cannabis_Consumers_12+"] = revised_percap_data["Cannabis_Consumers_12+"].str.replace(',','')
revised_percap_data
```

```python
revised_percap_data
```

```python
revised_percap_data["Cannabis_Consumption_Percentage"] = revised_percap_data["Cannabis_Consumption_Percentage"].str.replace('%','')
revised_percap_data
```

```python
revised_percap_data["Opioid_Deaths_Percentage"] = revised_percap_data["Opioid_Deaths_Percentage"].str.replace('%','')
revised_percap_data["Total_Drug_Deaths_Percentage"] = revised_percap_data["Total_Drug_Deaths_Percentage"].str.replace('%','')
revised_percap_data["Total_Drug_Deaths"] = revised_percap_data["Total_Drug_Deaths"].str.replace(',','')


revised_percap_data
```

```python
# Convert data types
revised_percap_data['Population'] = revised_percap_data['Population'].astype(int)
revised_percap_data['Cannabis Users 12+'] = revised_percap_data['Cannabis_Consumers_12+'].astype(int)
revised_percap_data['Cannabis Consumption Percentage'] = revised_percap_data['Cannabis_Consumption_Percentage'].astype(float)
revised_percap_data['Opioid Deaths'] = revised_percap_data['Opioid_Deaths'].astype(int)
revised_percap_data['Opioid Deaths Percentage'] = revised_percap_data['Opioid_Deaths_Percentage'].astype(float)
revised_percap_data['Total Drug Deaths'] = revised_percap_data['Total_Drug_Deaths'].astype(int)
revised_percap_data['Total Drug Deaths Percentage'] = revised_percap_data['Total_Drug_Deaths_Percentage'].astype(float)
revised_percap_data
```

```python
revised_percap_data.dtypes
```

```python
# Drop unnecessary columns
cleaned_revised_percap_data = revised_percap_data[['State', 'Population', 'Year', 'Cannabis Users 12+', 'Cannabis Consumption Percentage', 'Opioid Deaths', 'Opioid Deaths Percentage', 'Total Drug Deaths', 'Total Drug Deaths Percentage']]
cleaned_revised_percap_data
```

```python
cleaned_revised_percap_data.dtypes
```

```python
# Load legality
legality_df = pd.read_csv('../finalized_datasets/legality_merged.csv')
legality_df
```

```python
# Drop first column
legality_df.drop(columns=["Unnamed: 0"])
```

```python
legality_df.dtypes
```

```python
# Merge 
legality_merged_df = pd.merge(cleaned_revised_percap_data, legality_df, on=['State'], how='outer')
legality_merged_df
```
