```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Load file
us_opioid = pd.read_csv("all_us_opioid_deaths.csv")
us_opioid
```

```python
# Drop 'Unnamed:0'
us_opioid = us_opioid.drop(columns=["Unnamed: 0"])
us_opioid
```

```python
# load cannabis_df
cannabis_df = pd.read_csv("cannabis.csv")
cannabis_df
```

```python
# drop first column
cannabis_df = cannabis_df.drop(columns=["Unnamed: 0"])
cannabis_df
```

```python
# Data check
cannabis_df.dtypes
```

```python
# Convert to string
cannabis_df["Year"] = cannabis_df["Year"].astype(str)
cannabis_df
```

```python
# Check data type
cannabis_df.dtypes
```

```python
# Data check
us_opioid.dtypes
```

```python
us_opioid["Year"] = us_opioid["Year"].astype(str)
us_opioid
```

```python
# merge us_opioid and cannabis_df
us_opioid_cannabis = pd.merge(us_opioid, cannabis_df, on=['Year', 'State'], how='inner')
us_opioid_cannabis
```

```python
# multiply '12 or older estimate' values by 1000
us_opioid_cannabis['12 or Older Estimate'] = us_opioid_cannabis['12 or Older Estimate'].multiply(1000)
us_opioid_cannabis
```

```python
# load population data
us_pop = pd.read_csv("us_population.csv")
us_pop
```

```python
# Data check
us_pop.dtypes
```

```python
# Drop column
us_pop = us_pop.drop(columns=["Unnamed: 0"])
us_pop
```

```python
# Convert to string
us_pop["Year"] = us_pop["Year"].astype(str)
us_pop
```

```python
# merge us_opioid and cannabis_df
pop_op_can = pd.merge(us_opioid_cannabis, us_pop, on=['Year', 'State'], how='inner')
pop_op_can
```

```python
# Per capital cannabis users
percap_can = pop_op_can['12 or Older Estimate'] / pop_op_can['Population']
percap_can
```

```python
# Per Capita Opioid deaths
percap_op = pop_op_can['Number of Deaths'] / pop_op_can['Population']
percap_op
```

```python
# Create DataFrame
percap_data = pd.DataFrame({"State": pop_op_can["State"],
                            "Year": pop_op_can['Year'],
                            "Population": pop_op_can["Population"],
                            "Number of Cannabis Users": pop_op_can["12 or Older Estimate"],
                           "Number of Opioid Deaths": pop_op_can["Number of Deaths"],
                           "Per Capita Cannabis Use": percap_can,
                            "Per Capita Opioid Deaths": percap_op})
percap_data
```

```python
percap_data.to_csv("percap_data.csv")
```
