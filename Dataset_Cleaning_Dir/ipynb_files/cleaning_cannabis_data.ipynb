```python
import pandas as pd
import matplotlib as plt
```

```python
opioid_df = pd.read_csv("resources/2014-2020_Opioid_overdose_rates.csv")
opioid_df
```

```python
cannabis_df = pd.read_csv("resources/2014-2020_Cannabis_user_survey.csv")
cannabis_df.head(6)
```

```python
# Remove regional data
cannabis_df = cannabis_df.drop(cannabis_df.index[[1]])
cannabis_df
```

```python
# Alaska data
ak_df = cannabis_df[(cannabis_df["State"]=="Alaska")]
ak_df
```

```python
ak_df = ak_df[["12 or Older Estimate", "Year"]]
ak_df
```

```python
ak_df.info()
```

```python
# Convert  12 or older column to int64
ak_df["12 or Older Estimate"] = ak_df["12 or Older Estimate"].astype(int)
ak_df
```

```python
ak_df.info
```

```python
# Plot cannabis_df graph of 18+
import matplotlib.pyplot as plt
import pandas as pd
import time

plt.scatter(ak_df['Year'], ak_df['12 or Older Estimate'])
plt.ylabel('12 or Older Estimate')
plt.xlabel("Year")
plt.title('Cannabis Use by Year')
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
```
